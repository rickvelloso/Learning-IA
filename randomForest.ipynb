```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
base = pd.read_csv('insurance.csv', keep_default_na=False)
base = base.drop(columns=['Unnamed: 0'])
base.head()
```

```python
y = base.iloc[:,7].values
x = base.drop(base.columns[7], axis=1).values
x
```

```python
labelEncoder = LabelEncoder()
for i in range(x.shape[1]):
  if x[:,i].dtype == 'object':
    x[:,i] = labelEncoder.fit_transform(x[:,i])
x
```

```python
x_treinamento, x_test, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=12)
```

```python
model = RandomForestClassifier(random_state=1,max_depth=20, max_leaf_nodes=12, n_estimators=500)
model.fit(x_treinamento, y_treinamento)
```

```python
tree_index = 0
tree_to_visualize = model.estimators_[tree_index]
plt.figure(figsize=(10, 10))
plot_tree(tree_to_visualize, filled=True, feature_names=base.columns[:-1], class_names=True, rounded=True)
plt.show()
```

```python
previsoes = model.predict(x_test)
previsoes
```

```python
accuracy = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average='weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')
print(f'Acurácia: {accuracy}')
print(f'Precisão: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```

```python
report = classification_report(y_teste, previsoes)
print(report)
```
