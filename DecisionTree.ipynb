```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.tree import export_graphviz
import graphviz
```

```python
base = pd.read_csv('insurance.csv', keep_default_na=False)
base.head()
```

```python
#retira coluna Unnamed
base = base.drop(columns=['Unnamed: 0'])
base.head()
```

```python
y = base.iloc[:,7].values
x = base.drop(base.columns[7], axis=1).values
x
```

```python
labelEncoder = LabelEncoder()
for i in range(x.shape[1]):
  if x[:,i].dtype == 'object':
    x[:,i] = labelEncoder.fit_transform(x[:,i])
x
```

```python
x_treinamento, x_test, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=12)
```

Até aqui a base é a mesma do que foi feito no gaussina naive bayes.

```python
modelo = DecisionTreeClassifier(random_state=1,max_depth=8,max_leaf_nodes=6)
modelo.fit(x_treinamento,y_treinamento)
```

```python
dot_data = export_graphviz(modelo, out_file=None, filled=True, feature_names=base.columns[:-1])

graph = graphviz.Source(dot_data)
graph.render("decision_tree", format="png")
graph.view()
```

```python
previsoes = modelo.predict(x_test)
previsoes
```

```python
accuracy = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average='weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')
print(f'Acurácia: {accuracy}')
print(f'Precisão: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```

```python
report = classification_report(y_teste, previsoes)
print(report)
```
