```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
import numpy as np
```

```python
mtcars = pd.read_csv('mt_cars.csv')
mtcars.head()
```

```python
x = mtcars[['mpg', 'hp']].values
y = mtcars['cyl'].values
```

```python
num_rows = mtcars.shape[0]
print(f"O arquivo CSV tem {num_rows} linhas.")
```

```python
import math
print(math.sqrt(32))
```

Calculando a raiz quadrada de 32 (numero de linhas) para definir o valor de K.
Seguindo o padrão K = raiz de N.

```python
knn = KNeighborsClassifier(n_neighbors=5)
model = knn.fit(x, y)
model
```

```python
y_prev = model.predict(x)
print(y_prev)
```

```python
accuracy = accuracy_score(y, y_prev)
print(f"Acurácia: {accuracy}")
precision = precision_score(y, y_prev, average='weighted')
print(f"Precisão: {precision}")
recall = recall_score(y, y_prev, average='weighted')
print(f"Recall: {recall}")
f1 = f1_score(y, y_prev, average='weighted')
print(f"F1 Score: {f1}")
conf_matrix = confusion_matrix(y, y_prev)
print(f"Matriz de Confusão: \n {conf_matrix}")
```
